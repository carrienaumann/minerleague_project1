```python
school_data_to_load = "CN_school_disbursements_by_division.csv"
school_data = pd.read_csv(school_data_to_load)
school_data.head()
```

```python
#Henrico
labels = ['Administration', 'Transportation', 'Operations & Facilities','Technology', 'Instruction']
Values = [3.6, 5.56, 9.8, 5.6,75.5]
colors = ["red", "orange", "lightcoral", "lightskyblue"]
plt.pie(Values, labels=labels, colors=colors, autopct="%1.1f%%")
plt.axis("equal")
plt.title("Henrico", fontsize="15",fontweight="bold", color="blue", y=1.08)
plt.savefig("HenricoPie.png")
plt.show()
```
