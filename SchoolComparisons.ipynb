```python
# Impact of absenteeism on the graduation and drop out rates
%matplotlib notebook

# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Store the file name
myfile = "SchoolData.csv"
```

```python
# Read our Data file with the pandas library
df = pd.read_csv(myfile)
df.head()
```

```python
# Show graduation rate in bars and overlay a line chart to show absenteeism
ax=df.plot(kind="line", x='School Name', y='Absenteeism 10% or above', fontsize=7, color="red")
df.plot(kind="bar", x='School Name', y='Graduation Rate', facecolor="lightgreen",  fontsize=7, 
        position=0.5, figsize=(10,10),  ax=ax)
```

```python
# Save the chart to a png file
plt.savefig('Impact of Absenteeism on Graduation Rate')
```

```python
# Scatter Plot Graduation Rate vs Absenteeism
df.plot(kind="scatter", x="Absenteeism 10% or above", y="Graduation Rate", grid=True, figsize=(7,4),
              title="Graduation Rate vs Absenteeism", color='DarkBlue')
plt.show()
```

```python
# Save the chart to a png file
plt.savefig('Scatter - Graduation Rate vs Absenteeism.png')
```

```python
# Scatter Plot Dropout Rate vs Absenteeism
df.plot(kind="scatter", x="Absenteeism 10% or above", y="Dropout rate", grid=True, figsize=(7,4),
              title="Dropout rate vs Absenteeism", color='DarkBlue')
plt.show()
```

```python
# Save the chart to a png file
plt.savefig('Scatter - Dropout Rate vs Absenteeism.png')
```
